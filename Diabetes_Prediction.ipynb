```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
from sklearn.linear_model import LogisticRegression
```

```python
diabetes_data=pd.read_csv('diabetes.csv')
```

```python
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
#0 is non diabetics and 1 is diabetics 
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
diabetes_data.isnull().sum()
```

```python
features=diabetes_data.drop(columns='Outcome',axis=1) #Droping the column
target=diabetes_data['Outcome']
```

```python
print(features)
```

```python
print(target)
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(features)
```

```python
standardised_data=scaler.transform(features)
```

```python
standardised_data
```

```python
standardised_data=pd.DataFrame(standardised_data,columns=features.columns)
```

```python
standardised_data
```

```python
np.round(standardised_data.describe(),1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(standardised_data,target,test_size=0.2,random_state=2)
```

```python
print(features.shape,X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
classifier=LogisticRegression()
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print("AccuracyScore : ", training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print("AccuracyScore of testdata : ", testing_data_accuracy)
print("confusion_matrix\n",confusion_matrix(Y_test,X_test_prediction))
print("Precision",precision_score(Y_test,X_test_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf.fit(X_train, Y_train)
X_test_pred1=rf.predict(X_test)
print("confusion_matrix\n",confusion_matrix(Y_test,X_test_pred1))
print("Precision",precision_score(Y_test,X_test_pred1))
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf')
svm.fit(X_train, Y_train)

X_test_pred2 = svm.predict(X_test)

print("confusion_matrix\n", confusion_matrix(Y_test, X_test_pred2))
print("Precision", precision_score(Y_test, X_test_pred2))
```

```python
input_data=(5,166,72,19,175,25.8,0.587,51)

input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshaped)
print(std_data)
prediction=classifier.predict(std_data)
print(prediction)


if (prediction[0]==0):
    print("The person is not diabetic")
else:
    print("The person is diabetic")
```
